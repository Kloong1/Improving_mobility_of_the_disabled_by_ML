```python
import os, sys
```

```python
datapath = '/content/drive/Shareddrives/Project/수집데이터_전처리/'
```

```python
positions = os.listdir(datapath)

for position in positions:
  pospath = os.path.join(datapath, position)
  modes = os.listdir(pospath)

  print(f'--- {position} 데이터 분포 ---')
  sum = 0
  for mode in sorted(modes):
    modepath = os.path.join(pospath, mode)
    modecnt = len(os.listdir(modepath))
    sum += modecnt
    print(f'{mode}: {modecnt}')

  print(f'------- 총 {sum}개 -------\n')
```
