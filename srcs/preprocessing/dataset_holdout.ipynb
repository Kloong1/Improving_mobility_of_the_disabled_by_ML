```python
import os
import shutil
import random
```

# 경로 설정, Hold-out 비율 설정

```python
basicpath = '/content/drive/Shareddrives/Project/CNN_spectrogram_our_dataset/dataset'
```

```python
trainpath = os.path.join(basicpath, 'train')
valpath = os.path.join(basicpath, 'val')
testpath = os.path.join(basicpath, 'test')
```

```python
val_ratio = 0.1
test_ratio = 0.1
```

# 코드

```python
def _make_dir(dirpath):
  dirpath_split = dirpath.split(os.path.sep)
  new_path = '/'
  for directory in dirpath_split:
    new_path = os.path.join(new_path, directory)
    if not os.path.exists(new_path):
      os.mkdir(new_path)
```

```python
def seperate_dataset(position, mode, trainpath, valpath, testpath, val_ratio, test_ratio):
  trainpath = os.path.join(trainpath, position, mode)
  valpath = os.path.join(valpath, position, mode)
  testpath = os.path.join(testpath, position, mode)

  if not os.path.exists(valpath):
    _make_dir(valpath)
  
  if not os.path.exists(testpath):
    _make_dir(testpath)

  filepaths = [ os.path.join(trainpath, filename) for filename in os.listdir(trainpath) ]
  random.shuffle(filepaths)

  allcnt = len(filepaths)
  valcnt = int(allcnt * val_ratio)
  testcnt = int(allcnt * test_ratio)

  if valcnt == 0: valcnt = 1
  if testcnt == 0: testcnt = 1

  traincnt = allcnt - valcnt - testcnt

  if traincnt <= 0:
    traincnt = allcnt
    valcnt = 0
    testcnt = 0

  print(trainpath)
  print(f'all: {allcnt}  train: {traincnt}  val: {valcnt}  test: {testcnt}')

  print('val')
  for valfile in filepaths[:valcnt]:
    print(valfile)
    shutil.move(valfile, valpath)
  
  print('test')
  for testfile in filepaths[valcnt : valcnt + testcnt]:
    print(testfile)
    shutil.move(testfile, testpath)

  print('-----------------------------------------------------------------------------------------------------------------------\n')
```

# 실행

```python
positions = os.listdir(trainpath)
for position in positions:
  modes = os.listdir(os.path.join(trainpath, position))
  for mode in sorted(modes):
    seperate_dataset(position, mode, trainpath, valpath, testpath, val_ratio, test_ratio)
```
